```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canis_all.csv")
df.head()
```

```python
facebook = df['Facebook Follower #'].to_list()
twitter = df['X (Twitter) Follower #'].to_list()
youtube = df['YouTube Subscriber #'].to_list()
tiktok = df['TikTok Subscriber #'].to_list()
threads = df['Threads Follower #'].to_list()
```

```python
follower_frame = pd.DataFrame({
    'Facebook Follower #': facebook,
    'X (Twitter) Follower #': twitter,
    'YouTube Subscriber #': youtube,
    'TikTok Subscriber #': tiktok,
    'Threads Follower #': threads
})
```

```python
follower_frame.fillna(0, inplace=True)
follower_frame.head()
```

```python
follower_frame.to_csv('followers.csv')
```

```python
def find_string_rows(df):
    string_rows = {}
    for column in df.columns:
        # Check if each value in the column is an instance of str
        string_rows[column] = df[df[column].apply(lambda x: isinstance(x, str))].index.tolist()
    return string_rows

# Find rows with string values
string_rows = find_string_rows(df)
string_rows
```
