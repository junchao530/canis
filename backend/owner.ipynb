```python
df = pd.read_csv("canis_all.csv")
df.head()
```

```python
locations = df["Region of Focus"].to_list()
```

```python
owners = df['Entity owner (English)'].to_list()
companies = df['Name (English)'].to_list()
```

```python
ownership_dict = {}

for company, owner, location in zip(companies, owners, locations):
    if owner in ownership_dict:
        ownership_dict[owner][company] = location
    else:
        ownership_dict[owner] = {company: location}

for item, count in ownership_dict.items():
    print(f"{item}: {count}")

```

```python
import json

formatted_data = {}
for parent_company, companies in ownership_dict.items():
    formatted_data[parent_company] = {}

    for company, location in companies.items():
        if company not in formatted_data[parent_company]:
            formatted_data[parent_company][company] = location

# Writing the formatted data to a JSON file
with open('organization_data.json', 'w') as json_file:
    json.dump(formatted_data, json_file, indent=4)
```
