```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canis_all.csv")
df.head()
```

```python
influence_list = df["Region of Focus"].to_list()
```

```python
owners = df['Entity owner (English)'].to_list()
companies = df['Name (English)'].to_list()
```

```python
ownership_dict = {}

# Iterate through the lists
for company, owner in zip(companies, owners):
    # Check if the owner is already in the dictionary
    if owner in ownership_dict:
        # If yes, append the company to the list of owned companies
        ownership_dict[owner].append(company)
    else:
        # If not, create a new entry in the dictionary with the owner and the current company
        ownership_dict[owner] = [company]


for item, count in ownership_dict.items():
    print(f"{item}: {count} times")
```
